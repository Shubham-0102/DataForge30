```python
import pandas as pd

data = {
    "order_id": [101, 102, 103, 104, 105],
    "product": ["Laptop", "Phone", "Tablet", "Phone", "Laptop"],
    "quantity": [1, 2, 1, 3, 1],
    "price": [800, 500, 300, 500, 800],
    "order_date": [
        "2024-01-01",
        "2024-01-01",
        "2024-01-02",
        "2024-01-02",
        "2024-01-02"
    ]
}

df = pd.DataFrame(data)
df
```

```python
df["order_date"] = pd.to_datetime(df["order_date"])
df["total_price"] = df["quantity"] * df["price"]
df
```

```python
daily_sales = df.groupby("order_date").agg(
    total_orders=("order_id", "count"),
    total_revenue=("total_price", "sum"),
    avg_order_value=("total_price", "mean")
)

daily_sales
```

```python
with open("day8_daily_sales_report.txt", "w") as file:
    for date, row in daily_sales.iterrows():
        file.write(
            f"Date: {date.date()}\n"
            f"Total Orders: {row.total_orders}\n"
            f"Total Revenue: ${row.total_revenue}\n"
            f"Average Order Value: ${round(row.avg_order_value, 2)}\n\n"
        )

print("Report generated successfully")
```
