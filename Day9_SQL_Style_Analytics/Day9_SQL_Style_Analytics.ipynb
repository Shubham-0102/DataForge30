```python
import pandas as pd

data = {
    "order_id": [201, 202, 203, 204, 205, 206],
    "product": ["Laptop", "Phone", "Tablet", "Phone", "Laptop", "Tablet"],
    "quantity": [1, 2, 1, 3, 2, 2],
    "price": [900, 500, 300, 500, 900, 300],
    "order_date": [
        "2024-01-01",
        "2024-01-01",
        "2024-01-02",
        "2024-01-02",
        "2024-01-03",
        "2024-01-03"
    ]
}

df = pd.DataFrame(data)
df
```

```python
df["total_value"] = df["quantity"] * df["price"]
df
```

```python
revenue_per_product = (
    df.groupby("product")["total_value"]
    .sum()
    .sort_values(ascending=False)
)

revenue_per_product
```

```python
top_product = revenue_per_product.idxmax()
top_product
```

```python
df["order_date"] = pd.to_datetime(df["order_date"])

daily_revenue = (
    df.groupby("order_date")["total_value"]
    .sum()
)

daily_revenue
```

```python
high_value_orders = df[df["total_value"] > 1000]
high_value_orders
```

```python
with open("day9_sql_style_analysis.txt", "w") as f:
    f.write("DAY 9  SQL STYLE BUSINESS ANALYSIS\n\n")
    f.write("Revenue Per Product:\n")
    f.write(revenue_per_product.to_string())
    f.write("\n\nTop Selling Product:\n")
    f.write(top_product)
    f.write("\n\nDaily Revenue:\n")
    f.write(daily_revenue.to_string())
```
