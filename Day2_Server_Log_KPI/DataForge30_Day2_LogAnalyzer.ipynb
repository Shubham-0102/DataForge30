```python
# Count how many times each type appears
from collections import Counter

# Read the log file to ensure log_data is available
with open("server.log", "r") as f:
  log_data = f.read()

# Extract log types from each line
log_types = []
for line in log_data.split("\n"):
  if "[" in line and "]" in line:
    log_type = line.split("]")[0][1:]
    log_types.append(log_type)

counts = Counter(log_types)
counts
```

```python
from collections import Counter

# Read log file
with open("server.log", "r") as f:
    logs = f.readlines()

# Extract log types
log_types = [line.split("]")[0][1:] for line in logs]

# Count log types
counts = Counter(log_types)

# Save summary
with open("log_summary.txt", "w") as f:
    for log_type, count in counts.items():
        f.write(f"{log_type}: {count}\n")

print("log_summary.txt created successfully")
```
