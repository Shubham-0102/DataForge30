```python
import pandas as pd

data = {
    "Date": ["2025-12-30"]*5,
    "Employee": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Hours_Worked": [8, 7.5, 9, 8.5, 7],
    "Tasks_Completed": [5, 4, 6, 5, 3]
}

df = pd.DataFrame(data)
df.to_csv("daily_data.csv", index=False)
df
```

```python
df = pd.read_csv("daily_data.csv")
df
```

```python
total_employees = df.shape[0]
avg_hours = df["Hours_Worked"].mean()
total_tasks = df["Tasks_Completed"].sum()

report = f"""
Daily Report â€“ {df['Date'][0]}

Total Employees: {total_employees}
Average Hours Worked: {avg_hours:.2f}
Total Tasks Completed: {total_tasks}

Employee Details:
"""

for idx, row in df.iterrows():
    report += f"{row['Employee']}: Hours Worked={row['Hours_Worked']}, Tasks Completed={row['Tasks_Completed']}\n"
```

```python
with open("day6_daily_report.txt", "w") as f:
    f.write(report)

print("day6_daily_report.txt created")
```
